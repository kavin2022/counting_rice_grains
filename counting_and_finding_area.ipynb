# Getting area and the count of rice

```python
test = os.listdir('data/test')
train = os.listdir('data/train')
```

```python
# creating a data frame to store the output of watershed and contour method results
columns = ['name','method','total','area_avg','area_median','area']
df = pd.DataFrame(columns = columns)
```

###  Below code working
1. takes files from train and test directory
2. uses contour method and watershed method to get the area of the rice grain
3. stores the output image in train_solved and test_solved 
4. the functions return a dictionary containing
    1. total count
    2. area
    3. area_avg
    4. area_median

```python
# finding areas of rice grains through contour and watershed algorithm
for i in train:
    input_path = f'data/train/{i}'
    output_path_1 = f'data/train_solved/solved_contour_{i}'
    output_path_2 = f'data/train_solved/solved_watershed_{i}'
    result_1 = contour_method(input_path,output_path_1,True)
    result_2 = watershed_method(input_path,output_path_2,True)
    result_1['name'] = i
    result_2['name'] = i
    df = df.append(result_1,ignore_index = True)
    df = df.append(result_2,ignore_index = True)
    print(f'{i} : completed')
```

```python
# creating a data frame to store the output of watershed and contour method results
columns = ['name','method','total','area_avg','area_median','area']
df_test = pd.DataFrame(columns = columns)
```

```python
# finding areas of rice grains through contour and watershed algorithm
for i in test:
    input_path = f'data/test/{i}'
    output_path_1 = f'data/test_solved/solved_contour_{i}'
    output_path_2 = f'data/test_solved/solved_watershed_{i}'
    result_1 = contour_method(input_path,output_path_1,True)
    result_2 = watershed_method(input_path,output_path_2,True)
    result_1['name'] = i
    result_2['name'] = i
    df_test = df_test.append(result_1,ignore_index = True)
    df_test = df_test.append(result_2,ignore_index = True)
    print(f'{i} : completed')
```
